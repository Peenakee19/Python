**==> What is an Interpreter and a Compiler?**

someone who convert our code into machine language.

A Compiler translates the entire program into machine language at one time.

An Interpreter translates and executes the program line by line.

MAIN DIFFERENCES

Compiler

*  Translates whole program at once
*   Faster execution

*   Errors shown after full compilation
*   Creates a separate machine file

Example: C, C++

Interpreter

*  Translates line by line
*   Slower execution

*   Errors shown line by line
*   No separate file created

Example: Python, JavaScript
